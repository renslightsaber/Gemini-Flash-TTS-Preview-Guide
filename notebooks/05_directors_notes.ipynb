# 05. Director's Notes — 구조화된 프롬프트로 정교하게 연출하기

이 노트북에서 배우는 것:
- `style=` 한 줄보다 더 세밀한 제어가 필요할 때 쓰는 **Director's notes** 구조화 프롬프트
- `AUDIO PROFILE` / `SCENE` / `DIRECTOR'S NOTES` / `TRANSCRIPT` 섹션의 역할
- `TRANSCRIPT`만 실제로 발화되며, 나머지는 연출 지시라는 점
- `TRANSCRIPT`는 고정한 채 `DIRECTOR'S NOTES`만 바꿔서 딜리버리가 달라지는 걸 비교

## Director's notes 구조

`style=` 한 줄 지시보다 더 세밀하게 제어하고 싶을 때는, 아래처럼 구조화된 프롬프트 **블록 전체**를
`text` 인자로 넘기고 `style=None`으로 둡니다 (구조 자체가 이미 완성된 지시이므로 style 프리픽스를 붙이지 않습니다):

```
# AUDIO PROFILE: <화자 나이/성별/음색 한 줄>
## SCENE: <장면/분위기 한 줄>
### DIRECTOR'S NOTES
Style: <연기 톤>
Pacing: <속도/끊기>
Emphasis: <강조할 단어>
#### TRANSCRIPT
<실제로 읽을 문장 — 이 부분만 발화됨>
```

**핵심**: `AUDIO PROFILE` / `SCENE` / `DIRECTOR'S NOTES` 섹션은 모델이 어떻게 연기할지 "지시"만 하고,
실제로 소리 내어 읽는 부분은 `TRANSCRIPT` 섹션뿐입니다. `TRANSCRIPT` 안에는 03번 노트북에서 본
`[laughs]` 같은 인라인 태그도 그대로 넣을 수 있습니다.


````python
# 셋업: 상위 폴더(프로젝트 루트)를 sys.path에 추가해 gemini_tts 모듈을 import 합니다.
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import gemini_tts as gt
from IPython.display import Audio, display, Markdown

print("gemini_tts 로드 완료. 기본 모델:", gt.DEFAULT_MODEL)
````

## 1) 프리미엄 브랜드 내레이션 예시

전체 블록을 `text`로 그대로 넘기고 `style=None`으로 둡니다. 구조화된 프롬프트 자체가 이미 완성된 지시이기 때문입니다.

````python
director_block_v1 = """# AUDIO PROFILE: 30대 후반 여성, 낮고 우아한 저음
## SCENE: 프리미엄 브랜드 광고, 조용한 쇼룸에서 속삭이듯 소개하는 느낌
### DIRECTOR'S NOTES
Style: 절제되고 우아하게, 서두르지 않는 신뢰감 있는 톤
Pacing: 느리고 여유롭게, 문장 사이 짧은 호흡
Emphasis: "단 하나뿐인", "당신만을 위한"
#### TRANSCRIPT
단 하나뿐인 순간을 위해 준비했습니다. 이것은, 당신만을 위한 이야기입니다.
"""

wav, sr = gt.synthesize(director_block_v1, voice="Charon", style=None)
display(Audio(wav, rate=sr))
````

## 2) TRANSCRIPT는 그대로, DIRECTOR'S NOTES만 바꿔보기

같은 문장을 훨씬 빠르고 강조점이 다른 톤으로 연출을 바꿔서, 딜리버리가 어떻게 달라지는지 비교합니다.

````python
director_block_v2 = """# AUDIO PROFILE: 30대 후반 여성, 낮고 우아한 저음
## SCENE: 프리미엄 브랜드 광고, 조용한 쇼룸에서 속삭이듯 소개하는 느낌
### DIRECTOR'S NOTES
Style: 활기차고 설레는 톤, 살짝 들뜬 느낌
Pacing: 빠르고 경쾌하게, 문장 사이 거의 쉬지 않음
Emphasis: "지금", "바로"
#### TRANSCRIPT
단 하나뿐인 순간을 위해 준비했습니다. 이것은, 당신만을 위한 이야기입니다.
"""

wav2, sr2 = gt.synthesize(director_block_v2, voice="Charon", style=None)
display(Audio(wav2, rate=sr2))
````

---
**팁**
- `TRANSCRIPT` 섹션 안에도 `[laughs]`, `[whispers]` 같은 03번 노트북의 인라인 태그를 그대로 섞어 쓸 수 있습니다.
- `AUDIO PROFILE`은 보이스(`voice=`)가 이미 정해주는 음색과 별개로, 화자의 배경 설정을 문맥으로 더 보강해주는 역할을 합니다.
- CLI에서는 `python cli.py --prompt-file director.md --voice Charon`처럼 이 블록을 파일로 저장해 재사용할 수 있습니다.
