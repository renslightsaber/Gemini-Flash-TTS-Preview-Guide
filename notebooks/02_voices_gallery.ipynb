# 02. Voices Gallery — 30종 프리셋 보이스 둘러보기

이 노트북에서 배우는 것:
- `gt.VOICES` 전체 30종 보이스 표 출력
- `gt.FEMALE_VOICES` / `gt.MALE_VOICES` / `gt.voices_for()`로 성별 힌트 기반 부분집합 고르기
- 큐레이션된 소수(6~8개) 보이스로 같은 문장을 합성해 나란히 청음

**주의**: 30개를 전부 합성하면 API 호출이 많아지므로, 이 노트북은 일부러 소수 서브셋만 돕니다. 전체를 청음하고 싶으면 아래 `curated` 리스트를 직접 늘려보세요.

```python
# 셋업: 상위 폴더(프로젝트 루트)를 sys.path에 추가해 gemini_tts 모듈을 import 합니다.
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import gemini_tts as gt
from IPython.display import Audio, display, Markdown

print("gemini_tts 로드 완료. 기본 모델:", gt.DEFAULT_MODEL)
```

## 1) 전체 30종 보이스 표

보이스 이름 → 스타일 descriptor(공식 표기). 성별([F]/[M])은 공식 라벨이 아니라 `gt.FEMALE_VOICES` / `gt.MALE_VOICES` 기반의 대략적인 청감 분류입니다.

```python
# gt.VOICES 전체를 표 형태로 출력합니다.
print(f"프리셋 보이스 {len(gt.VOICES)}종 (이름 · 스타일 · 성별힌트):\n")
for i, (name, style) in enumerate(gt.VOICES.items(), 1):
    hint = "F" if name in gt.FEMALE_VOICES else ("M" if name in gt.MALE_VOICES else " ")
    print(f"  {i:>2}. {name:<14} {style:<14} [{hint}]")
```

## 2) 큐레이션된 서브셋 청음

여성 보이스 몇 개 + 남성 보이스 몇 개를 골라 **같은 문장**으로 합성합니다. 톤(스타일 descriptor)이 다르면 같은 문장도 인상이 어떻게 달라지는지 비교해보세요.

`voices_for("female")`, `voices_for("male")`로 성별 힌트 목록 전체를 얻을 수 있습니다 (정답이 아니라 참고용 분류입니다).

```python
# 청음할 동일 문장.
sample_text = "이 제품은 정말 특별한 순간을 선물해 드립니다."

# 큐레이션된 보이스 서브셋 (여성 4 + 남성 4). 이 리스트를 늘리면 더 많은 보이스를 오디션할 수 있습니다.
curated = (gt.FEMALE_VOICES[:4] + gt.MALE_VOICES[:4])
print("이번에 청음할 보이스:", curated)
```

```python
# 각 보이스로 같은 문장을 합성해 라벨과 함께 재생합니다.
for voice in curated:
    style_descriptor = gt.VOICES.get(voice, "?")
    gender_hint = "F" if voice in gt.FEMALE_VOICES else ("M" if voice in gt.MALE_VOICES else "?")
    wav, sr = gt.synthesize(sample_text, voice=voice)
    display(Markdown(f"**{voice}** — {style_descriptor} (성별힌트: {gender_hint})"))
    display(Audio(wav, rate=sr))
```

## 3) 더 넓게 오디션하고 싶다면

`curated` 리스트에 `gt.VOICES.keys()`의 다른 이름들을 추가하면 30종 전체(또는 원하는 만큼)를 같은 방식으로 오디션할 수 있습니다. 다만 호출 수만큼 API 사용량이 늘어나니 필요한 만큼만 추가하세요.

예: `curated = list(gt.VOICES.keys())`  ← 전체 30종
