# 04. Style Instructions — 자연어 지시로 톤 통째로 바꾸기

이 노트북에서 배우는 것:
- `synthesize(text, style=...)`의 `style`은 `"<style>: <text>"` 형태로 텍스트 앞에 붙는 자연어 톤 지시입니다 (`gt.build_contents` 참고).
- **같은 문장**을 고정하고 `style`만 바꿔가며 합성해서, 얼마나 다른 인상을 주는지 비교합니다.
- 이건 CLI의 `python cli.py "..." --style "..."`와 동일한 레버입니다.

```python
# 셋업: 상위 폴더(프로젝트 루트)를 sys.path에 추가해 gemini_tts 모듈을 import 합니다.
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import gemini_tts as gt
from IPython.display import Audio, display, Markdown

print("gemini_tts 로드 완료. 기본 모델:", gt.DEFAULT_MODEL)
```

## 고정 문장

아래 한 문장을 style만 바꿔가며 4가지 톤으로 합성합니다.

```python
# 이 문장은 모든 스타일 실험에서 동일하게 유지합니다.
fixed_text = "지금 바로 만나보세요. 특별한 혜택이 기다리고 있습니다."
```

### 스타일 1) 신뢰감 있고 차분한 내레이션

```python
wav, sr = gt.synthesize(fixed_text, voice="Charon", style="신뢰감 있고 차분하게 (내레이션)")
display(Audio(wav, rate=sr))
```

### 스타일 2) 긴급 프로모션처럼 힘있고 빠르게

```python
wav, sr = gt.synthesize(fixed_text, voice="Kore", style="긴급 프로모션처럼 힘있고 빠르게")
display(Audio(wav, rate=sr))
```

### 스타일 3) 다정하고 속삭이듯 부드럽게

```python
wav, sr = gt.synthesize(fixed_text, voice="Vindemiatrix", style="다정하고 속삭이듯 부드럽게")
display(Audio(wav, rate=sr))
```

### 스타일 4) 밝고 통통 튀는 MZ 톤

```python
wav, sr = gt.synthesize(fixed_text, voice="Sadachbia", style="밝고 통통 튀는 MZ 톤으로")
display(Audio(wav, rate=sr))
```

---
**참고**: `style`은 보이스(voice)와 독립적인 레버입니다. 같은 보이스라도 style을 바꾸면 톤이 달라지고, 같은 style이라도 보이스를 바꾸면 음색 자체가 달라집니다. 두 레버를 조합해 원하는 톤을 찾아보세요.
