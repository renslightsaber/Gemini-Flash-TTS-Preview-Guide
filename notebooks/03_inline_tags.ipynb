# 03. Inline Tags — 문장 속 대괄호 태그로 비언어적 연기 넣기

이 노트북에서 배우는 것:
- 문장 **중간**에 `[laughs]`, `[sighs]`, `[whispers]` 같은 대괄호 태그를 넣으면 모델이 그 지점에서 해당 비언어적 소리/톤을 연기합니다.
- 미리 정의된 태그(`gt.INLINE_TAGS`) 외에도 **커스텀 태그**(`[very fast]`, `[like a cartoon dog]` 등)도 자유 서술로 동작합니다(`gt.CUSTOM_TAG_EXAMPLES`).
- 태그는 텍스트 안에 그대로 섞어 쓰면 되며, `style=` 인자와 함께 써도 됩니다.

```python
# 셋업: 상위 폴더(프로젝트 루트)를 sys.path에 추가해 gemini_tts 모듈을 import 합니다.
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import gemini_tts as gt
from IPython.display import Audio, display, Markdown

print("gemini_tts 로드 완료. 기본 모델:", gt.DEFAULT_MODEL)
```

## 1) 사용 가능한 태그 레퍼런스

`gt.INLINE_TAGS`는 자주 쓰는 감정/비언어 태그 예시이고, `gt.CUSTOM_TAG_EXAMPLES`는 자유 서술 커스텀 태그의 예시입니다. 이 목록에 없는 태그도 자연어로 서술하면 대체로 동작합니다.

```python
print("자주 쓰는 인라인 태그:")
print(gt.INLINE_TAGS)

print("\n커스텀 태그 예시(자유 서술):")
print(gt.CUSTOM_TAG_EXAMPLES)
```

## 2) `[laughs]` — 웃음을 문장 중간에 삽입

화자가 놀라며 웃는 반응을 연기하도록 유도합니다.

```python
text_laughs = "정말요? [laughs] 믿을 수가 없어요!"
wav, sr = gt.synthesize(text_laughs, voice="Aoede")
display(Audio(wav, rate=sr))
```

## 3) `[sighs]` — 문장 시작에 한숨

지친 하루를 표현하는 한숨으로 문장을 시작합니다.

```python
text_sighs = "[sighs] 오늘 하루 정말 길었어요."
wav, sr = gt.synthesize(text_sighs, voice="Charon")
display(Audio(wav, rate=sr))
```

## 4) `[whispers]` — 속삭이는 톤

문장 전체를 속삭이듯 조용한 톤으로 유도합니다.

```python
text_whispers = "[whispers] 조용히 해, 다들 자고 있어."
wav, sr = gt.synthesize(text_whispers, voice="Enceladus")
display(Audio(wav, rate=sr))
```

## 5) `[excited]` — 흥분/기쁨 톤

성취의 순간을 신나는 톤으로 표현합니다.

```python
text_excited = "[excited] 우리가 해냈어요!"
wav, sr = gt.synthesize(text_excited, voice="Puck")
display(Audio(wav, rate=sr))
```

## 6) 커스텀 태그 — `[very fast]` (속도 지시)

`gt.INLINE_TAGS`에 없는 자유 서술 태그도 동작합니다. 여기서는 말하는 속도를 지시해봅니다.

```python
text_fast = "[very fast] 마감이 오늘까지예요 서둘러요!"
wav, sr = gt.synthesize(text_fast, voice="Fenrir")
display(Audio(wav, rate=sr))
```

---
**팁**: 태그는 문장 어디에나(맨 앞/중간/맨 뒤) 넣을 수 있고, 여러 개를 한 문장에 섞어 쓸 수도 있습니다. 다만 태그가 너무 많으면 자연스러움이 떨어질 수 있으니 1~2개로 시작하는 걸 추천합니다.
