## API 키(.env) 준비

- `gt.synthesize()`를 호출하면 내부에서 `python-dotenv`로 프로젝트 루트의 `.env` 파일을 자동으로 읽어
  `GEMINI_API_KEY`를 로드합니다. 이 노트북에서 직접 키를 다루지 않아도 됩니다.
- `.env`가 없다면, 프로젝트 루트의 `.env.example`을 복사해 `.env`를 만들고
  `GEMINI_API_KEY=AIza...` 형태로 키를 넣어주세요 (https://aistudio.google.com 에서 발급).
- 키가 없으면 `synthesize()` 호출 시 친절한 `RuntimeError` 메시지가 안내해줍니다.


```python
# 셋업: 상위 폴더(프로젝트 루트)를 sys.path에 추가해 gemini_tts 모듈을 import 합니다.
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import gemini_tts as gt
from IPython.display import Audio, display, Markdown

print("gemini_tts 로드 완료. 기본 모델:", gt.DEFAULT_MODEL)
```

## 1) 가장 단순한 합성

기본 보이스는 `Kore` (Firm 톤)입니다. `style`을 주지 않으면 텍스트를 그대로 자연스럽게 읽습니다.

```python
# 짧은 한국어 문장을 기본 보이스(Kore)로 합성합니다.
text = "안녕하세요, 오늘도 좋은 하루 보내세요."

wav, sr = gt.synthesize(text, voice="Kore")

print(f"샘플레이트: {sr} Hz, 길이: {len(wav) / sr:.2f}초")
display(Audio(wav, rate=sr))
```

## 2) `style=`로 톤 지시 추가하기

`style`은 자연어 지시를 텍스트 앞에 `"<style>: <text>"` 형태로 붙여 모델에 전달합니다 (`gt.build_contents` 참고). 같은 문장이라도 톤이 달라지는지 들어보세요.

```python
# 같은 문장, style만 추가해서 톤을 바꿔봅니다.
wav2, sr2 = gt.synthesize(text, voice="Kore", style="밝고 따뜻하게")

display(Audio(wav2, rate=sr2))
```
